## Layers are made of neurons
* Fully connected when each neuron links to
 all neurons in the previous layer
*  A neuron in a linear layer:
1.    A neuron in a linear layer:
 Performs a linear operation using all
 neurons from the previous layer
2.   Has N+1 parameters: N from inputs and 1
 for the bias





## Parameters and model capacity

```python
model = nn.Sequential(
    nn.Linear(8,4),
    nn.Linear(4,2)
)
```

Screenshot 2025-05-16 032327.png

 * First layer has 4 neurons, each neuron has
 8+1 parameters. 9 times 4 = 36 parameters

 * Second layer has 2 neurons, each neuron
 has 4+1 parameters. 5 times 2 = 10
 parameters

* 36 + 10 = 46 learnable parameters

```python
total = 0
for parameter in model.parameters():
  total += parameter.numel()
print(total)
```

 * .numel() : returns the number of elements
 in the tensor
